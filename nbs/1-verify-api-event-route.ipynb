```python
import requests
```

```python
path = "/api/events/"
base_url = "http://localhost:8002"
endpoint = f"{base_url}{path}"
endpoint
```

```python
from requests import Response


response: Response = requests.get(endpoint)
```

```python
response.ok
```

```python
if response.ok:
    data = response.json()
    print(data, type(data), data.get('items'))
```
